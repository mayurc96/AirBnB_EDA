### Knowing the dataset



The dataset contains 65282 rows X 20 columns.
1. `CustKey`: A unique identifier for each customer.
2. `DateKey`: A unique identifier or date value representing a specific date in the dataset (e.g., sales date, order date).
3. `Discount Amount`: The amount of discount applied to the sales.
4. `Invoice Date`: The date when the invoice was generated or issued.
5. `Invoice Number`: A unique identifier for each invoice.
6. `Item Class`: A category or classification of the item.
7. `Item Number`: A unique identifier for each item.
8. `Item`: The name or description of the item being sold.
9. `Line Number`: A numerical identifier for each line item in the invoice or order.
10. `List Price`: The original or standard price of the item before any discounts or promotions.
11. `Order Number`: A unique identifier for each order.
12. `Promised Delivery Date`: The date on which the delivery of the order is promised or expected.
13. `Sales Amount`: The total amount of sales (after applying discounts) for a specific item or order.
14. `Sales Amount Based on List Price`: The sales amount calculated based on the list price without discounts.
15. `Sales Cost Amount`: The cost amount associated with the sales (e.g., the cost of goods sold).
16. `Sales Margin Amount`: The profit margin amount for the sales (Sales Amount - Sales Cost Amount).
17. `Sales Price`: The actual price at which the item was sold after applying discounts.
18. `Sales Quantity`: The quantity of items sold in a specific transaction or order.
19. `Sales Rep`: The sales representative or agent associated with the sale.
20. `U/M`: Unit of Measure, representing the unit in which the sales quantity is measured (e.g., pieces, units).


```python
# Create new column as total_amount
sales_df['total_amount'] = (sales_df['Sales Price'] * sales_df['Sales Quantity']).round(2)
```

```python
sales_df
```

```python

# Convert 'DateKey' to datetime format
sales_df['DateKey'] = pd.to_datetime(sales_df['DateKey'])

# Set 'DateKey' as the index
sales_df.set_index('DateKey', inplace=True)

# Resample to calculate weekly, monthly, and daily totals
daily_total = sales_df['total_amount'].resample('D').sum()
monthly_total = sales_df['total_amount'].resample('M').sum()
yearly_total = sales_df['total_amount'].resample('Y').sum()

# Print the results
print("daily Total Sales:")
print(daily_total)

print("\nMonthly Total Sales:")
print(monthly_total)

print("\nYearly Total Sales:")
print(yearly_total)
```

```python
# Group daily totals by days of the week and sort in ascending order
daily_totals_by_day = daily_total.groupby(daily_total.index.day_name()).sum()
sorted_daily_totals = daily_totals_by_day.reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

# Print the results
print("Daily Total Sales by Day (Ascending Order):")
print(sorted_daily_totals)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
sorted_daily_totals.plot(kind='line', color='red')
plt.title('Daily Total Sales by Day (Ascending Order)')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

#Identify top customers with the highest sales or most frequent purchases.

```python
# including 'CustKey', 'total_amount', and 'Invoice Number'

# Convert 'CustKey' to a string, if it's not already
sales_df['CustKey'] = sales_df['CustKey'].astype(str)

# Calculate total sales amount per customer
customer_total_sales = sales_df.groupby('CustKey')['total_amount'].sum()

# Calculate purchase frequency per customer
customer_purchase_frequency = sales_df['CustKey'].value_counts()

# Merge the calculated metrics into a single DataFrame
customer_metrics = pd.DataFrame({
    'Total Sales Amount': customer_total_sales,
    'Purchase Frequency': customer_purchase_frequency
})

# Sort the DataFrame by total sales amount in descending order
top_customers_by_sales = customer_metrics.sort_values(by='Total Sales Amount', ascending=False)

# Sort the DataFrame by purchase frequency in descending order
top_customers_by_frequency = customer_metrics.sort_values(by='Purchase Frequency', ascending=False)

# Print the top customers by sales and by purchase frequency
print("Top Customers by Sales:")
print(top_customers_by_sales.head())

print("\nTop Customers by Purchase Frequency:")
print(top_customers_by_frequency.head())
```

## Identify top-selling products based



```python
# including 'Item Number', 'Item', 'Sales Quantity', and 'total_amount'

# Calculate total sales quantity per product
product_total_quantity = sales_df.groupby(['Item Number', 'Item'])['Sales Quantity'].sum()

# Calculate total sales amount per product
product_total_sales = sales_df.groupby(['Item Number', 'Item'])['total_amount'].sum()

# Merge the calculated metrics into a single DataFrame
product_metrics = pd.DataFrame({
    'Total Sales Quantity': product_total_quantity,
    'Total Sales Amount': product_total_sales
}).reset_index()

# Sort the DataFrame by total sales amount in descending order
top_selling_products = product_metrics.sort_values(by='Total Sales Amount', ascending=False).head(10)

# Print the top-selling products
print("Top Selling Products:")
print(top_selling_products)
```

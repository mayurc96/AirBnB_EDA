## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

# 1)Top 5 host with maximum number of bookings
# Calculate the number of bookings per host
host_bookings = airbnb_df.groupby(['host_id', 'host_name', 'neighbourhood_group']).size().reset_index(name='booking_count')

# Sort the hosts by booking count in descending order
sorted_hosts = host_bookings.sort_values('booking_count', ascending=False).reset_index(drop=True)

# Select the top 5 hosts
top_5_hosts = sorted_hosts.head(5)

# Display the table
top_5_hosts
```

```python
# 2)Number of bookings per neighbourhood group
neighbourhood_group_bookings = airbnb_df['neighbourhood_group'].value_counts()
neighbourhood_group_bookings
```

```python
# 3)Number of bookings per room type
room_type_bookings = airbnb_df['room_type'].value_counts()
room_type_bookings
```

```python
# 4)Popular neighbourhoods within neighbourhood group

# Group the data by neighbourhood group and neighbourhood, and count the bookings
popular_neighbourhoods = airbnb_df.groupby(['neighbourhood_group', 'neighbourhood']).size().reset_index(name='booking_count')

# Sort the neighbourhoods within each group by booking count in descending order
popular_neighbourhoods_sorted = popular_neighbourhoods.sort_values(['neighbourhood_group', 'booking_count'], ascending=[True, False])

# Select the top three popular neighbourhoods within each group
top_neighbourhoods = popular_neighbourhoods_sorted.groupby('neighbourhood_group').head(3)
top_neighbourhoods
```

```python
# 5)Availability per neighborhood group
availability_per_group = airbnb_df.groupby('neighbourhood_group')['availability_365'].median()
availability_per_group
```

```python
# 6)Availability per neighbourhood group and room type
availability_per_group_room = airbnb_df.groupby(['neighbourhood_group', 'room_type'])['availability_365'].median()

# Reshape the data for plotting
availability_per_group_room = availability_per_group_room.unstack()
availability_per_group_room
```

```python
# 7)Price vs neighborhood group
price_vs_neighborhood_group = airbnb_df.groupby('neighbourhood_group',as_index=False)['price'].median()
price_vs_neighborhood_group
```

```python
# 8)Room Type vs Price
room_type_vs_price = airbnb_df.groupby('room_type',as_index=False)['price'].median()
room_type_vs_price
```

```python
# 9)Price vs Room Type for each Neighbourhood Group
price_vs_room_neighbourhood = airbnb_df.groupby(['neighbourhood_group', 'room_type'])['price'].median().reset_index()
price_vs_room_neighbourhood
```

```python
# 10)Relation between minimum night and room type
# Group the data by room type and calculate the median of minimum nights
min_nights_room_type = airbnb_df.groupby('room_type')['minimum_nights'].median().reset_index()
min_nights_room_type
```

```python
# 11)Relation between minimum night and room type for every neighbourhood group
# Group the data by room type and neighbourhood group and calculate the median of minimum nights
min_nights_room_type_neighbourhood = airbnb_df.groupby(['neighbourhood_group', 'room_type'])['minimum_nights'].median().reset_index()
min_nights_room_type_neighbourhood
```

```python
# 12)Host with most listed properties
max_host_list = airbnb_df.groupby(['host_id', 'host_name'])['calculated_host_listings_count'].mean().reset_index().sort_values(by='calculated_host_listings_count',ascending=False).head(10)
max_host_list
```
